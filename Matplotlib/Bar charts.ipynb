```python
import matplotlib.pyplot as plt
import numpy as np
import random
from matplotlib import style
```

```python
classes=['PY','R','AI','ML','DS']
class1=[30,10,20,25,10]
class2=[45,5,20,20,10]
class3=[35,5,30,15,15]
```

```python
plt.barh(classes,class1)
plt.show()
```

```python
plt.bar(classes,class1)
plt.show()
```

```python
plt.figure(figsize=(16,9))
style.use("ggplot")
plt.bar(classes,class1,width=0.5,align="edge",color='y',edgecolor="b",alpha=0.5,linestyle="--")
plt.title("Bar chart of IAIP class")
plt.xlabel("classes")
plt.ylabel("No of students")
plt.legend(["Class-1 Students"])
plt.show()
```

```python
plt.figure(figsize=(16,9))
style.use("ggplot")
plt.bar(classes,class1,width=0.3,align="edge",color='y',edgecolor="k",alpha=0.5,linestyle="-")
plt.bar(classes,class2,width=0.3,align="edge",color='b',edgecolor="k",alpha=0.5,linestyle="-")
plt.bar(classes,class3,width=0.3,align="edge",color='g',edgecolor="k",alpha=0.5,linestyle="-")
plt.title("Bar chart of IAIP class")
plt.xlabel("classes")
plt.ylabel("No of students")
plt.legend(["Class-1 Students"])
plt.show()
```

```python
plt.figure(figsize=(16,9))
style.use("ggplot")
width=0.3
class_index=np.arange(len(classes))
plt.bar(class_index,class1,width,color='y',edgecolor="k",alpha=0.5,linestyle="-")
plt.bar(class_index + width,class2,width,color='g',edgecolor="k",alpha=0.5,linestyle="-")
plt.bar(class_index + width*2,class3,width,color='r',edgecolor="k",alpha=0.5,linestyle="-")
plt.xticks(class_index +width,classes,rotation=20)
plt.title("Bar chart of IAIP class")
plt.xlabel("classes")
plt.ylabel("No of students")
plt.legend(["Class-1 Students","Class-2 Students","Class-3 Students"])
plt.show()
```

```python
plt.figure(figsize=(16,9))
style.use("ggplot")
width=0.15
plt.bar(classes,class3,width=-0.3*1.5,align="edge",color='y',edgecolor="k",alpha=0.8,linestyle="-")
plt.bar(classes,class2,width=0.3*1.5,align="edge",color='g',edgecolor="k",alpha=0.8,linestyle="-")
plt.bar(classes,class1,width=0.3,align="center",color='b',edgecolor="k",alpha=1,linestyle="-")
plt.title("Bar chart of IAIP class")
plt.xlabel("classes")
plt.ylabel("No of students")
plt.legend(["Class-1 Students","Class-2 Students","Class-3 Students"])
plt.show()
```

```python
plt.figure(figsize=(16,9))
style.use("ggplot")
width=0.3
class_index=np.arange(len(classes))
plt.barh(class_index,class1,width,color='y',edgecolor="k",alpha=0.5,linestyle="-")
plt.barh(class_index + width,class2,width,color='g',edgecolor="k",alpha=0.5,linestyle="-")
plt.barh(class_index + width*2,class3,width,color='b',edgecolor="k",alpha=0.5,linestyle="-")
plt.yticks(class_index +width,classes,rotation=20)
plt.title("Bar chart of IAIP class")
plt.xlabel("classes")
plt.ylabel("No of students")
plt.legend(["Class-1 Students","Class-2 Students","Class-3 Students"])
plt.show()
```
