```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import style
import pandas
```

```python
classes=["Python","R","Machine learning","Artificial inteligence","Data science"]
class1=[45,15,35,25,30]
```

```python
plt.pie([1])
plt.show()
```

```python
explode=[0,0,0.1,0,0]
plt.pie(class1,labels=classes,explode=explode,autopct="%0.2f%%",shadow=True,radius=1.5)
plt.show()
```

```python
explode=[0,0,0.1,0,0]
plt.pie(class1,labels=classes,explode=explode,autopct="%0.2f%%",shadow=True,radius=2,frame=True,pctdistance=0.6)
plt.legend(loc=3)
plt.show()
```
